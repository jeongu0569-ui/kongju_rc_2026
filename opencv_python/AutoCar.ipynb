```python
import time 
from pop import Pilot

car = Pilot.AutoCar()
```

```python
#좌(-1) ~ 우(1)

car.steering = -1   #좌
time.sleep(1)
car.steering = 1   #우
time.sleep(1)
car.steering = 0   #중앙
```

```python
#전진 및 후진

car.backward(50)
time.sleep(1)
car.forward(50)
car.stop()
```

```python
#조이스틱 조정

car.joystick()
```

```python
car.camPan(0)
time.sleep(1)
car.camPan(180)
car.camTilt(100)
time.sleep(1)
car.camTilt(0)
```

```python
value = car.getGyro()
print(value)
```

```python
#자이로센서(좌표값계산)

from IPython.display import clear_output
import time

while True:
    clear_output(wait=True)
    value = car.getGyro()
    value2 = car.getAccel()
    print(value)
    print(value2)
    time.sleep(0.1)
```

```python
#이미지 가져오기
import cv2
filename = 'img.jpg'
image = cv2.imread(filename)


#높낮이 및 채널 추출
height, width, channel = image.shape
print(f"width {width}, height {height}, channel {channel}")


#이미지 화면 표시 기능
cv2.imshow("image", image)


#흑백 변경 및 외각선 추출
image2 = cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
canny = cv2.Canny(image,100,200)

cv2.imshow("image1",image2)
cv2.imshow("image2",canny)


#이미지 색상 반전하기
b, g, r = cv2.split(image) #opencv에서는 색상 순서를 bgr로 한다

image_rgb = cv2.merge((r,g,b)) # 일상생활에서 쓰이는 순서
cv2.imshow("image_rgb", image_rgb)
cv2.imwrite("img_rgb.jpg", image_rgb)


#이미지 압축 JPEG 저장하기 
cv2.imwrite(
    "img_jpeg.jpg",
    image,
    [cv2.IMWRITE_JPEG_QUALITY, 1]
)
img_jpeg = cv2.imread("img_jpeg.jpg")
cv2.imshow("img_jpeg", img_jpeg)
```

```python
cam = Util.gstrmer(width=640, height=480, fps=30, flip=0)
camera = cv2.VideoCapture(cam, cv2.CAP_GSTREAMER)

if not camera.isOpened():
    print("Not found camera")

for _ in range(120):
    ret, frame = camera.read()
    if not ret:
        print(ret)
        continue
    cv2.imshow("soda", frame)
    
camera.release()
```
